```python
import json

import matplotlib.pyplot as plt
import numpy as np
```

```python
books = {
    "shakes": "Shakespeare Coll. (1590–1616)",  # Active writing period
    "victorian": "Victorian Poetry (1800–1900)",  # Victorian poetry era
    "dikens": "C. Dickens Coll. (1836–1870)",  # Dickens' career (Pickwick Papers ~ Death)
    # "crime": "Crime and Punishment (~1866)",  # Published in 1866
    "warpeace": "War and Peace (–1869)",  # Published serially 1865–1869
    "stephen": "S. King Coll. (1980–2000)",  # First book *Carrie* in 1974, still ongoing
    "potter": "Harry Potter (1997–2007)",  # Series publication period
    "modern": "Contemporary Poetry (2000–)",  # Open-ended, ongoing works
    "articles": "Academic Papers (2000–)",  # Open-ended, continuously published
}
```

```python
pos_dist_dict_book = {}
for book in books:
    with open(f"results/distance/{book}.json", "r") as file:
        pos_dist_dict_book[book] = json.load(file)

plt.figure(dpi=200)
plt.barh(books.values(), [pos_dist_dict_book[book]["all"] for book in books])
plt.xlabel("Mean distance to [0, 1, 0]")
plt.gca().invert_yaxis()
plt.savefig(f"results/figs/mean_dist.pdf", bbox_inches="tight", pad_inches=0.05)
plt.show()
```

```python
classes = {
    "intj": "Interjection (Wow, Ha)",
    "noun": "Noun (apple, car)",
    "verb": "Verb (run, eat)",
    "adj": "Adjective (happy, tall)",
    "adv": "Adverb (quickly, softly)",
    "sconj": "Subord. Conj. (if, as)",
    "aux": "Auxiliary Verb (is, has)",
    "adp": "Adposition (in, on)",
    "pron": "Pronoun (she, they)",
    "cconj": "Coord. Conj. (and, or)"
}
```

```python
for book in books:
    plt.figure(dpi=200)
    plt.barh(pos_dist_dict_book[book].keys(), pos_dist_dict_book[book].values())
    plt.title(books[book])
    plt.xlabel("Mean distance to [0, 1, 0]")
    plt.xlim(0, 0.07)
    plt.savefig(f"figs/{book}_dist.pdf", bbox_inches="tight", pad_inches=0.05)
    plt.show()
```
